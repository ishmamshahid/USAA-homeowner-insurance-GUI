### The first page (page 8 of the pdf)

```python
import PyPDF2
import re
import numpy as np
import pandas as pd

# Function to extract specific lines from a given page number
def extract_specific_lines_from_page(pdf_path, page_number, start_line, end_line):
    with open(pdf_path, 'rb') as file:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(file)
        
        # Check if the page number is within the range
        if page_number < len(pdf_reader.pages):
            # Extract text from the specific page
            page = pdf_reader.pages[page_number]
            text = page.extract_text()
            
            # Split the text into lines
            lines = text.split('\n')
            
            # Get the desired lines (adjusting for zero-based indexing)
            if len(lines) >= end_line:
                return lines[start_line-1:end_line]
            else:
                return "Page does not contain enough lines."
        else:
            return "Page number is out of range."

# Example usage
pdf_path = 'D:\\LSU\\Research\\wind_ins\\USAA-134022151\\Rate-Rule Attachments\\LA Rate Pages - Part 1.pdf'   # Replace with your file path
page_number = 7  # Page 8 in zero-indexed format
start_line = 12
end_line = 13
specific_lines = extract_specific_lines_from_page(pdf_path, page_number, start_line, end_line)
#print("\n".join(specific_lines))

# Convert list of strings into a single string and split into individual values
data_values = ' '.join(specific_lines).split()

# Convert list of strings to a NumPy array of floats
array_data = np.array(data_values, dtype=float)

filtered_arr = array_data[(array_data % 1 != 0)] 
df = pd.DataFrame(filtered_arr.reshape(-1, 10))
rating_factor_table_1st_page = df.applymap(lambda x: np.floor(x * 10000) / 10000)
rating_factor_table_1st_page
```
