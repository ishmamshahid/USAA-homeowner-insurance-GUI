### Rating Area  - Rating Area Group, pdf part 2, pg 115

```python
import PyPDF2
import re
import numpy as np
import pandas as pd

# Function to extract specific lines from a given page number
def extract_specific_lines_from_page(pdf_path, page_number, start_line, end_line):
    with open(pdf_path, 'rb') as file:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(file)
        
        # Check if the page number is within the range
        if page_number < len(pdf_reader.pages):
            # Extract text from the specific page
            page = pdf_reader.pages[page_number]
            text = page.extract_text()
            
            # Split the text into lines
            lines = text.split('\n')
            
            # Get the desired lines (adjusting for zero-based indexing)
            if len(lines) >= end_line:
                return lines[start_line-1:end_line]
            else:
                return "Page does not contain enough lines."
        else:
            return "Page number is out of range."

# Example usage
pdf_path = 'D:\\LSU\\Research\\wind_ins\\USAA-134022151\\Rate-Rule Attachments\\LA Rate Pages - Part 2.pdf'   # Replace with your file path
page_number = 114 
start_line = 13
end_line = 15
specific_lines = extract_specific_lines_from_page(pdf_path, page_number, start_line, end_line)
#print("\n".join(specific_lines))

```

```python
# Convert list of strings into a single string and split into individual values
data_values = ' '.join(specific_lines).split()
data_values = combined_string = "".join(data_values)
single_digit = data_values[:27]

def add_spaces(text):
    result = ''
    for i in range(0, len(text), 2):
        if i + 1 < len(text):
            result += text[i] + ' ' + text[i+1]
        else:
            result += text[i]
        if i + 2 < len(text):
            result += ' '
    return result.strip()

single_digit = add_spaces(single_digit)

# Replace "A " with "A" to remove the space after "A"
single_digit = single_digit.replace('A ', 'A')

single_digit
```

```python
two_digit_1st_pg = data_values[27:]

# Adding a space after every two characters
two_digit_1st_pg = ' '.join([two_digit_1st_pg [i:i+2] for i in range(0, len(two_digit_1st_pg), 2)])

two_digit_1st_pg 
```

```python
first_page = single_digit + " " + two_digit_1st_pg

def string_to_list(input_string):
    return input_string.split()

first_page = string_to_list(first_page)

# Function to create a dataframe with odd and even positioned elements in separate columns
def create_dataframe(input_list):
    odd_position_elements = input_list[::2]  # Get odd positioned elements (index starts from 0)
    even_position_elements = input_list[1::2]  # Get even positioned elements
    # Fill shorter list with None to match the length of the longer list
    if len(odd_position_elements) > len(even_position_elements):
        even_position_elements.append(None)
    # Create dataframe
    df = pd.DataFrame({
        'Rating Area': odd_position_elements,
        'Rating Area Group': even_position_elements
    })
    return df

# Create the dataframe
first_page = create_dataframe(first_page)
first_page
```
